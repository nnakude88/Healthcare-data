```python
import pandas as pd
import plotly.express as px
import seaborn as sns

# Load the dataset
hc = pd.read_excel("healthcare_dataset.xlsx")

# Create a vibrant color palette
colors = sns.color_palette("husl", 8).as_hex()

# Create an interactive histogram for Age
fig = px.histogram(hc, x='Age', nbins=20, title='Age Distribution',
                   color_discrete_sequence=[colors[0]])

# Show the interactive plot
fig.show()
```

```python
import plotly.express as px

# Prepare the data by counting the values
medical_condition_data = hc['Medical Condition'].value_counts().reset_index()
medical_condition_data.columns = ['Medical Condition', 'Count']  # Rename columns appropriately

# Create an interactive bar chart for Medical Condition distribution
fig = px.bar(medical_condition_data,
             x='Medical Condition', y='Count', 
             title='Medical Condition Distribution',
             labels={'Medical Condition': 'Medical Condition', 'Count': 'Count'},
             color='Count',  # Color by the count of each medical condition
             color_continuous_scale=px.colors.sequential.Plasma)  # Color scale

# Show the interactive plot
fig.show()
```

```python
import plotly.express as px

# Prepare the data
gender_counts = hc['Gender'].value_counts().reset_index()
gender_counts.columns = ['Gender', 'Count']

# Create an interactive pie chart for Gender distribution
fig = px.pie(gender_counts, 
             names='Gender', 
             values='Count', 
             title='Gender Distribution',
             color_discrete_sequence=colors)

# Show the interactive plot
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Load the dataset
hc = pd.read_excel("healthcare_dataset.xlsx")

# Count the number of admissions per hospital
admissions_count = hc['Hospital'].value_counts().reset_index()
admissions_count.columns = ['Hospital', 'Number of Admissions']

# Get the top 10 hospitals with the most admissions
top_10_hospitals = admissions_count.nlargest(10, 'Number of Admissions')

# Create a bar chart for the top 10 hospitals
fig_bar = px.bar(top_10_hospitals, x='Hospital', y='Number of Admissions',
                  title='Top 10 Hospitals with Most Admissions',
                  labels={'Hospital': 'Hospital', 'Number of Admissions': 'Total Admissions'},
                  color='Number of Admissions',  # Color by number of admissions
                  color_continuous_scale=px.colors.sequential.Viridis)

# Show the interactive plot
fig_bar.show()
```

```python
import pandas as pd
import plotly.express as px

# Load the dataset
hc = pd.read_excel("healthcare_dataset.xlsx")

# Count the number of admissions per insurance provider
insurance_count = hc['Insurance Provider'].value_counts().reset_index()
insurance_count.columns = ['Insurance Provider', 'Number of Admissions']

# Get the top 10 insurance providers with the most admissions
top_10_insurance = insurance_count.nlargest(10, 'Number of Admissions')

# Create a bar chart for the top 10 insurance providers
fig_bar = px.bar(top_10_insurance, x='Insurance Provider', y='Number of Admissions',
                  title='Top 10 Insurance Providers with Most Admissions',
                  labels={'Insurance Provider': 'Insurance Provider', 'Number of Admissions': 'Total Admissions'},
                  color='Number of Admissions',  # Color by number of admissions
                  color_continuous_scale=px.colors.sequential.Viridis)

# Show the interactive plot
fig_bar.show()
```

```python
import pandas as pd
import plotly.express as px

# Load the dataset
hc = pd.read_excel("healthcare_dataset.xlsx")

# Create a new DataFrame for the counts of Medical Condition by Test Results
count_data = hc.groupby(['Test Results', 'Medical Condition']).size().reset_index(name='Count')

# Create a bar chart
fig_bar = px.bar(count_data, 
                  x='Test Results', 
                  y='Count', 
                  color='Medical Condition',  # Color by Medical Condition
                  title='Medical Condition by Test Results',
                  labels={'Test Results': 'Test Results', 'Count': 'Count'},
                  barmode='group',  # Group bars for better comparison
                  color_discrete_sequence=px.colors.qualitative.Set2)

# Show the interactive plot
fig_bar.show()
```
